# 14.1 결정 트리 분류기 훈련

결정 트리 분류기를 훈련 - 사이킷런의 DecisionTreeClassifier 사용

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import datasets

# 데이터를 로드
iris = datasets.load_iris()
X,Y = iris.data, iris.target

# 결정 트리가 분류기를 만듬

decisiontree = DecisionTreeClassifier(random_state=0)

# 모델을 훈련
model= decisiontree.fit(X,Y)
```

```python
X
```

```python
Y
```

결정 트리 학습기는 노드에서 불순도가 가장 크게 감소하는 결정 규칙을 찾음. 불순도를 측정하는 방법은 많이 있지만 DecisionTreeClassifier는 기본적으로 지니 불순도를 사용. 

사이킷런의 DecisionTreeClassifier 클래스의 동작 방식은 다른 학습기와 비슷. fit 메서드를 사용해 모델을 훈련한 뒤 이 모델을 사용해 샘플의 클래스를 예측


```python
# 새로운 샘플을 생성

observation = [[5,4,3,2]]

# 샘플의 클래스를 예측
model.predict(observation)
```

```python
# 샘플의 클래스 예측 확률
# 세개의 클래스에 대한 예측 확률을 확인

model.predict_proba(observation)
```

```python
# 마지막으로 다른 불순도 지표를 사용하고 싶다면 criterion 매개변수를 사용

# 엔트로피를 사용해 결정 트리 분류기를 훈련.
decisiontree_entropy = DecisionTreeClassifier(criterion='entropy', random_state=0)

# 모델을 훈련
model_entropy = decisiontree_entropy.fit(X,Y)
```

```python
model_entropy.predict(X)
```

# 14.2 결정 트리 회귀 훈련

결정 트리를 사용한 회귀 모델을 훈련 -- 사이킷런의 DecisionTreeRegressor 를 사용


```python
from sklearn.tree import DecisionTreeRegressor
from sklearn import datasets

# 데이터를 로드하고 두 개의 특성만 선택
boston= datasets.load_boston()

X= boston.data[:,:2]
Y=boston.target

# 결정 트리 회귀모델을 만듬

decisiontree = DecisionTreeRegressor(random_state=0)

model = decisiontree.fit(X,Y)
```

결정 트리 회귀는 결정 트리 분류와 작동 방식이 비슷. 지니 불순도나 엔트로피를 감소하는 대신 기본적으로 얼마나 MSE를 감소시키는지에 따라 분할.

```python
# 새로운 샘플 생성.
observation=[[0.02,16]]

model.predict(observation)
```

DecisionTreeClassifier처럼 criterion 매개변수를 사용해 분할 품질의 측정 방식을 선택. 예를 들어 분할할 때 'MAE'가 감소되는 트리 생성.

```python
decisiontree=DecisionTreeRegressor(random_state=0,criterion='mae')

model = decisiontree.fit(X,Y)

model.predict(observation)
```

```python
# 라이브러리를 임포트합니다.
import pydotplus
from sklearn.tree import DecisionTreeClassifier
from sklearn import datasets
from IPython.display import Image
from sklearn import tree

# 데이터를 로드합니다.
iris = datasets.load_iris()
features = iris.data
target = iris.target

# 결정 트리 분류기를 만듭니다.
decisiontree = DecisionTreeClassifier(random_state=0)

# 모델을 훈련합니다.
model = decisiontree.fit(features, target)

# DOT 데이터를 만듭니다.
dot_data = tree.export_graphviz(decisiontree,
                                out_file=None,
                                feature_names=iris.feature_names,
                                class_names=iris.target_names)

# 그래프를 그립니다.
graph = pydotplus.graph_from_dot_data(dot_data)

# 그래프를 출력합니다.
Image(graph.create_png())
```

```python
# pdf로 만들기

graph.write_pdf('iris.pdf')

graph.write_png('iris.png')
```

# 14.4 랜덤 포레스트 분류기 훈련하기


랜던함 결정 트리를 많이 사용해 분류 모델을 훈련 <- 사이킷런의 RandomForestClassifier를 사용해 랜덤 포레스트 분류 모델 훈련

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import datasets

iris = datasets.load_iris()
X,Y= iris.data, iris.target

# 랜덤 포레스트 분류기 객체를 만듬. 
randomforest = RandomForestClassifier(random_state=0, n_jobs=-1)

model= randomforest.fit(X,Y)

```

```python
X.shape
```

```python
X
```

결정 트리의 일반적인 문제는 훈련 데이터에 너무 가깝게 맞추려는 경향이 있음(즉, 과대적합) 이런 이유 때문에 랜덤 포레스트라 불리는 앙상블 방법이 널리 사용. 랜덤 포레스트는 많은 결정 트리를 훈련하지만 각 트리는 부트스트랩 샘플을 사용.(즉, 원본 샘플 수와 동일하게 중복을 포함해 랜덤하게 샘플을 뽑음). 또 각 노드는 최적의 분할을 결정할 때 특성의 일부만 사용. 이 랜덤한 결정 트리의 숲이 투표하여 예측 클래스를 결정.

RandomForestClassifier는 DecisionTreeClassifier와 작동 방식이 유사



```python
# 새로운 샘플을 만듬
observation = [[5, 4 , 3, 2]]

# 샘플 클래스를 예측
model.predict(observation)

```

RandomForestClassifier는 DecisionTreeClassifier와 같은 매개변수가 많음. 예를들어 분할 품질의 측정 방식을 바꿀 수 있음

```python
# 엔트로피를 사용해 랜덤 포레스트 분류기 객체를 만듬.

randomforest_entropy = RandomForestClassifier(criterion='entropy', random_state= 0)

# 모델 훈련

model_entropy = randomforest_entropy.fit(X,Y)
```

개별 결정 트리가 아니므로 RandomForestClassifier에만 있거나 특별히 중요한 매개변수가 있음. 첫째, max_feaures 매개변수는 각 노드에서 
사용할 특성의 최대 개수를 결정. 이 매개변수는 정수(특성의 수), 실수(특성 개수 비율), sqrt(특성 개수의 제곱근),를 입력할 수 있다. max_features의 기본값은 auto로 sqrt와 동작이 같음. 둘째,bootstrap 매개변수는 트리에 사용할 샘플을 중복을 허용한 샘플링 만들지(기본값) 아닐지를 결정. 셋째 n_estimators는 랜덤 포레스트엣허 만들 결정 트리 개수를 지정 . 레시피 10.4 에서 n_estimators를 하이퍼 파라미터로 취급하여 트리 개수가 증가함에 따라 평가 지표에 미치는 영향을 확인 . 마지막으로 많은 결정 트리 모델을 만들 때 효율적이므로 n_jobs = -1 로 설정하여 모든 코어를 사용한느 것이 좋음.

# 14.5 랜덤 포레스트 회귀 훈련

랜덤 포레스트를 사용한 회귀 모델을 훈련 - RandomForestRegressor 

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import load_boston

boston = load_boston()
X,Y= boston.data[:,0:2], boston.target

RFregr =RandomForestRegressor(random_state = 0,n_jobs=-1)

model = RFregr.fit(X,Y)
```

RandomForestClassifier에서 처럼 RandomForestRegressor도 몇가지 중요한 매개 변수가 있음.

1. max_features는 각 노드에서 사용할 최대 특성 개수를 지정. 기본값은 전체 특성 개수

2. bootstrap은 중복을 허용한 샘플링 여부를 지정. 기본값은 True

3. n_estimators는 사용할 결정 트리 개수를 지정. 기본값은 10

# 14.6 랜덤 포레스트에서 중요한 특성 구분


랜덤 포레스트 모델에서 가장 중요한 특성 -특성 중요도를 계산하고 시각화

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn import datasets

iris = datasets.load_iris()

X= iris.data
Y= iris.target

randomforest = RandomForestClassifier(random_state = 0, n_jobs=-1)

model = randomforest.fit(X,Y)

# 특성 중요도를 계산 !!
importances = model.feature_importances_


## 특성 중요도를 내림차순으로 정렬
indices = np.argsort(importances)[::-1]

# 정렬된 특성 중요도에 따라 특성의 이름을 나열

names = [iris.feature_names[i] for i in indices]
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
plt.figure()

plt.title("Feature Importance")
plt.bar(range(X.shape[1]), importances[indices])

plt.xticks(range(X.shape[1]),names, rotation=90)

plt.show()
```

```python
X.shape
```

```python
indices
```

```python
importances  # 특성 중요도의 합은 전체 1
```

```python
iris.feature_names
```

```python
iris.target_names
```

랜덤 포레스트의 특성 중요도는 결정 트리의 중요도를 특성별로 평균하여 특성 중요도를 구함. 최종적으로 특성 중요도가 합이 1이 되도록 정규화, 단일 결정 트리에서 만든 특성 중요도와 랜덤 포레스트의 특성 중요도를 비교하면 랜덤 포레스트가 한 특성에 치우치지 않고 데이터의 다양한 면을 바라봄.

```python
# 결정 트리 분류기를 만듬
from sklearn.tree import DecisionTreeClassifier
decisiontree = DecisionTreeClassifier(random_state = 0)

model =decisiontree.fit(X,Y)

# 랜덤 포레스트와 결정 트리의 특성 중요도를 비교

fig,ax = plt.subplots()

rects1 = ax.bar(np.arange(X.shape[1])-0.25,
               randomforest.feature_importances_,0.5,
               label = 'Random Forest')

rects2 = ax.bar(np.arange(X.shape[1])+0.25,model.feature_importances_,0.5,label="Decision Tree")

plt.xticks(range(X.shape[1]),names, rotation=90)
plt.legend()
plt.show()
```

# 14.7 랜덤 포레스트에서 중요한 특성 선택

랜덤 포레스트에서 특성 선택을 수행 <- 중요한 특성을 구별하고 가장 중요한 특성만 사용해 모델을 다시 훈련

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn import datasets
from sklearn.feature_selection import SelectFromModel

# 데이터를 로드
iris = datasets.load_iris()
X= iris.data
Y= iris.target

randomforest = RandomForestClassifier(random_state = 0 , n_jobs =-1)

# 특성 중요도가 임곗값보다 크거나 같은 특성으로 객체 생성.
selector = SelectFromModel(randomforest, threshold = 0.3)

# selector를 사용해 새로운 특성 행렬 생성
features_important = selector.fit_transform(X,Y)

# 가장 중요한 특성을 사용해 랜덤 포레스트 모델을 훈련

model = randomforest.fit(features_important,Y)
```

모델 특성 개수를 감소시켜야 할 때가 있음. 예를 들어 모델의 분산을 감소시키거나 가장 중요한 특성만 사용해 모델을 이해하기 쉽게 만들어야 하는 경우.

사이킷런에서는 두 단계의 워크플로를 사용해 줄어든 특성으로 모델을 만들 수 있음. 먼저 모든 특성을 사용해 랜덤 포레스트 모델 훈련 후 그다응ㅁ 중요한 특성만 포함된 새로운 특성 행렬 생성. 해결에서 SelectFromModel 클래스를 사용해 threshold 값보다 중요도가 크거나 같은 특성만 포함된 특성 행렬 생성. 마지막으로 이 특성 만을 이용해 새로운 모델 훈련 

이것에 대한 두가지 단점

1. 원-핫 인코딩딘 순서가 없는 범주형 특성의 특성 중요도는 여러 개의 이진 특성으로 희석. 둘째, 상관관계가 높은 특성의 중요도는 양쪽 특성에 고루 분산되는 것이 ㅇ아닌 한 특성에 집중.
